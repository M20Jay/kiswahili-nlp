**Purpose:** Classify Kiswahili environmental text by UNEP Strategic Objective

**Model:** bert-base-multilingual-cased (mBERT) — supports 104 languages including Kiswahili  

**Tools:** HuggingFace Transformers · MLflow · Python

## Section 1: Importing Libraries 

```python
# ── Section 1: Import Libraries ──────────────────────────────────
# transformers → HuggingFace library that gives us access to mBERT
# pipeline     → a simplified way to use any HuggingFace model
#                in just 2 lines of code
# mlflow       → experiment tracking — records every run we do

from transformers import pipeline
import mlflow
import pandas as pd

print("✅ Libraries imported successfully")
```

## Section 2: Defining Kiswahili Environmental Sentences

```python
# ── Section 2: Define Kiswahili Environmental Sentences ──────────
# These are real Kiswahili sentences about environmental topics
# Each sentence connects to a UNEP Strategic Objective
# SO1 = Climate Stability
# SO2 = Biodiversity
# SO3 = Pollution and Waste

sentences = [
    "Misitu inakatwa kwa kasi kubwa barani Afrika",
    "Mabadiliko ya tabianchi yanaathiri wakulima wadogo",
    "Hewa chafu inadhuru afya za watu mijini",
    "Viumbe vingi vya porini viko hatarini kutoweka",
    "Plastiki nyingi zinatupwa baharini na kudhuru samaki",
    "Ukame unaoongezeka unasababisha njaa Afrika Mashariki",
    "Nishati mbadala inaweza kupunguza uchafuzi wa hewa",
    "Misitu ya mvua inahifadhi bioanuwai ya kipekee"
]

# What each sentence means in English:
meanings = [
    "Forests are being cut down rapidly across Africa",
    "Climate change is affecting small-scale farmers",
    "Air pollution is harming the health of urban people",
    "Many wildlife species are at risk of extinction",
    "Much plastic is dumped in the ocean harming fish",
    "Increasing drought is causing hunger in East Africa",
    "Renewable energy can reduce air pollution",
    "Rainforests preserve unique biodiversity"
]

print(f"✅ {len(sentences)} Kiswahili sentences ready")
print("\nSentences loaded:")

for i, (sw,en) in enumerate(zip(sentences,meanings),1):
    print(f"{i}.{sw}")
    print(f"\n {en}")
```

## Section 3: Loading mBERT Model

```python

# ── Section 3: Load mBERT Model ──────────────────────────────────
# We are loading mBERT from HuggingFace
# mBERT = multilingual BERT
# Trained on Wikipedia in 104 languages including Kiswahili
# Task: zero-shot-classification
#       means we classify without training
#       we just give it labels and it figures it out

print("Loading mBERT model...")
print("This may take 1-2 minutes on first run")
print("HuggingFace will download the model weights")
print("Approximately 700MB — saved locally after first download")
print()

classifier = pipeline(
    "zero-shot-classification",
    model = "facebook/bart-large-mnli"
)
print("✅ Model loaded successfully")
```

## Section 4: Defining UNEP Labels and Classifying

```python
# ── Section 4: Define Labels and Classify ────────────────────────
# We define the UNEP Strategic Objective labels
# The model will score each sentence against these labels
# and return which label fits best

# These are the categories we want to classify into
labels = [
    "climate change and carbon emissions",
    "biodiversity loss and wildlife extinction", 
    "pollution and plastic waste"
]

# What each label maps to:
# Label 1 → UNEP Strategic Objective 1 — Climate Stability
# Label 2 → UNEP Strategic Objective 2 — Biodiversity
# Label 3 → UNEP Strategic Objective 3 — Pollution and Waste

print("✅ Labels defined:")

for i, label in enumerate(labels,1):
    print(f"SO{i}: {label}")
```

## Section 5: Classify Each Sentence

```python
# ── Section 5: Classify Each Sentence ───────────────────────────
# Now we run each Kiswahili sentence through the model
# classifier() takes:
#   → the sentence (what to classify)
#   → candidate_labels (what categories exist)
# It returns:
#   → labels ranked by confidence score
#   → scores between 0 and 1
#   → higher score = stronger match

print("Classifying Kiswahili sentences...\n")

results = []
for i, (sentence,meaning) in enumerate(zip(sentences,meanings),1):
    # Run the sentence through the model
    output = classifier(sentence, candidate_labels=labels)

    # The top prediction is always at position 0
    top_label = output['labels'][0]
    top_score = output['scores'][0]

    # Store the result
    results.append({
        'sentence': sentence,
        'meaning': meaning,
        'predicted_label': top_label,
        'confidence': round(top_score, 4)
    })


    print(f"Sentence {i}: {sentence}")
    print(f"English:   {meaning}")
    print(f"Predicted: {top_label}")
    print(f"Confidence: {top_score:.2%}")
    print()
print("✅ Classification complete") 
```

## Section 6: Results Summary Table

```python
# ── Section 6: Results Summary Table ─────────────────────────────
# Convert results to a pandas DataFrame
# DataFrame = a table with rows and columns
# like an Excel spreadsheet in Python

df_results = pd.DataFrame(results)

# Add a column mapping prediction to SO number
so_mapping = {
    "climate change and carbon emissions": "SO1",
    "biodiversity loss and wildlife extinction": "SO2",
    "pollution and plastic waste": "SO3"
}

df_results['strategic_objective'] = df_results['predicted_label'].map(so_mapping)

print("Results Summary:")
print(df_results[['meaning', 'strategic_objective', 'confidence']].to_string(index=False))
```

## Section 7: Track Experiment with MLflow

```python
# ── Section 7: MLflow Experiment Tracking ────────────────────────
# MLflow records this experiment run
# So we can compare it with future fine-tuned versions
# Every run is logged with:
# → parameters (what settings we used)
# → metrics (how well it performed)

import mlflow

# Set the experiment name
# All runs for this project go under this name
mlflow.set_experiment("kiswahili-nlp")

# Start a run — like opening a new page in a lab notebook
with mlflow.start_run(run_name="zero-shot-prototype"):

    # Log parameters — what we used
    mlflow.log_param("model", "facebook/bart-large-mnli")
    mlflow.log_param("approach", "zero-shot-classification")
    mlflow.log_param("language", "Kiswahili")
    mlflow.log_param("num_sentences", len(sentences))
    mlflow.log_param("num_labels", len(labels))

    # Calculate metrics — how well it performed
    avg_confidence = df_results['confidence'].mean()
    correct = 3  # sentences 4, 5, 8 were correct
    accuracy = correct / len(sentences)

    # Log metrics
    mlflow.log_metric("average_confidence", round(avg_confidence, 4))
    mlflow.log_metric("correct_predictions", correct)
    mlflow.log_metric("accuracy", round(accuracy, 4))

    print("✅ MLflow experiment logged")
    print(f"   Experiment: kiswahili-nlp")
    print(f"   Run name:   zero-shot-prototype")
    print(f"   Avg confidence: {avg_confidence:.2%}")
    print(f"   Accuracy: {accuracy:.2%}")
    print()
    print("To view results run in terminal:")
    print("   mlflow ui")
    print("Then open: http://localhost:5000")
```
